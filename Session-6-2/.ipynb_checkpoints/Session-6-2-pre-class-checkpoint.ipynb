```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm

def CI_95(N, p): 
    dist = norm(loc=p, scale=np.sqrt(p*(1-p)/N))
    margin = (dist.interval(0.95)[1]-p)*100
    print('Error margin for {:.1f} is: {:.1f}%'.format(p, margin))
    x = np.linspace(0,1,1000)
    plt.plot(x, dist.pdf(x), label=N)

CI_95(1000,0.1)
plt.show()
```

We are considering the number of success of a yes/no question, and binomial distribution as it expresses the number of success from N Bernoulli trials.

```python
for i in [100,1000,10000]:
    print("Error margins for {:d}".format(i))
    for _ in [0.2,0.4,0.6,0.8]:
        CI_95(i, _)
    print ("\n")
        
plt.show()
```
