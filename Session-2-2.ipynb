## Plot histograms and pdfs

```python
# import statistic and plot libraries
from scipy.stats import gamma, norm
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

### Gamma distributions

```python
# define the range of x for plot
x = np.linspace(0, 16, 100)

plt.plot(x, gamma.pdf(x, a=1.5, scale=1), 'b-')
plt.plot(x, gamma.pdf(x, a=3, scale=1.2), 'r-')
plt.title("pdf of gamma distribution")
function_1 = mpatches.Patch(color='b', label="α=1.5, β=1")
function_2 = mpatches.Patch(color='r', label="α=3, β=1.2")
plt.legend(handles=[function_1, function_2])
plt.show()

plt.hist(gamma.rvs(a=3, scale=1.2, size=1000), density=True, bins=list(range(0, 16)))
plt.plot(x, gamma.pdf(x, a=3, scale=1.2), 'r-')
plt.title("1000 sample histogram v.s. pdf function")
plt.show()
```

### Normal distributions

```python
# define the range of x for plot
x = np.linspace(-12, 12, 100)

plt.plot(x, norm.pdf(x, scale=3), 'b-')
plt.plot(x, norm.pdf(x, loc=-4, scale=2), 'r-')
plt.title("pdf of normal distribution")
function_1 = mpatches.Patch(color='b', label='μ=0, σ=3')
function_2 = mpatches.Patch(color='r', label='μ=-4, σ=2')
plt.legend(handles=[function_1, function_2])
plt.show()

plt.hist(norm.rvs(loc=-4, scale=2, size=1000), density=True, bins=list(range(-12, 12)))
plt.plot(x, norm.pdf(x, loc=-4, scale=2), 'r-')
plt.title("1000 sample histogram v.s. pdf function")
plt.show()
```
